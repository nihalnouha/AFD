```python
import numpy as np
import matplotlib.pyplot as plt

# Données
Y1 = np.array([7, 9, 8, 10, 11, 12, 13])
Y2 = np.array([26, 28, 27, 23, 25, 24, 29])
Y3 = np.array([84, 84, 81, 81, 80, 79, 78])

# Centrage des données
X1 = Y1 - np.mean(Y1)
X2 = Y2 - np.mean(Y2)
X3 = Y3 - np.mean(Y3)

# Matrice des données centrées
X = np.column_stack((X1, X2, X3))

# Matrice de variance totale
V = np.dot(X.T, X) / 7

# Groupes
Q = np.repeat(["p1", "p2", "p3"], [2, 2, 3])

# Matrice barycentre des classes centres
G = np.array([[np.mean(X1[Q=="p1"]), np.mean(X1[Q=="p2"]), np.mean(X1[Q=="p3"])],
              [np.mean(X2[Q=="p1"]), np.mean(X2[Q=="p2"]), np.mean(X2[Q=="p3"])],
              [np.mean(X3[Q=="p1"]), np.mean(X3[Q=="p2"]), np.mean(X3[Q=="p3"])]])

# Matrice interclasses
B = np.dot(G.T, G) / 3

# Fonction pour calculer la matrice intraclasse
def V_prime(Y1, Y2, Y3):
    X1 = Y1 - np.mean(Y1)
    X2 = Y2 - np.mean(Y2)
    X3 = Y3 - np.mean(Y3)
    X = np.column_stack((X1, X2, X3))
    return np.dot(X.T, X) / len(Y1)

# Calcul des matrices intra-classes
V1 = V_prime([7, 9], [26, 28], [84, 84])
V2 = V_prime([8, 10], [27, 23], [81, 81])
V3 = V_prime([11, 12, 13], [25, 24, 29], [80, 79, 78])

# Matrice intra-classes
W = (2 * V1 + 2 * V2 + 3 * V3) / 7

# Calcul V^-1 * B
VB = np.linalg.inv(V) @ B

# Vecteurs propres
valeurs_propres, vecteurs_propres = np.linalg.eig(VB)
vecteur_propre_1 = vecteurs_propres[:, 0]
vecteur_propre_2 = vecteurs_propres[:, 1]

# Normalisation
norme_vecteur_propre_1 = np.sqrt(vecteur_propre_1.T @ np.linalg.inv(V) @ vecteur_propre_1)
norme_vecteur_propre_2 = np.sqrt(vecteur_propre_2.T @ np.linalg.inv(V) @ vecteur_propre_2)
vect_unitaire_1 = vecteur_propre_1 / norme_vecteur_propre_1
vect_unitaire_2 = vecteur_propre_2 / norme_vecteur_propre_2

# Projections des individus
CC1 = X @ np.linalg.inv(V) @ vect_unitaire_1
CC2 = X @ np.linalg.inv(V) @ vect_unitaire_2

# Projection des centres de gravité
C1 = G @ np.linalg.inv(V) @ vect_unitaire_1
C2 = G @ np.linalg.inv(V) @ vect_unitaire_2

# Graphique des résultats
plt.figure()
plt.plot(CC1, CC2, 'o', label="Individus")
plt.plot(C1, C2, 'o', label="Centres de gravité")
for i, txt in enumerate(Q):
    plt.annotate(txt, (CC1[i], CC2[i]))
plt.xlabel("Axe discriminant 1")
plt.ylabel("Axe discriminant 2")
plt.title("Projection sur le plan discriminant")
plt.legend()
plt.show()
```
